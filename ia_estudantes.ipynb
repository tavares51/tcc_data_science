**Um pouco sobre Burnout, ansiedade e depressão**
<br>
**Burnout**: A Síndrome de Burnout é uma condição psicológica decorrente de estresse crônico no ambiente de trabalho, caracterizada por exaustão emocional, despersonalização e redução da realização pessoal. Ela resulta da percepção de desequilíbrio entre as demandas profissionais e os recursos disponíveis, afetando negativamente a saúde mental e a produtividade do indivíduo.

**Ansiedade**: A ansiedade é uma resposta emocional natural a situações de ameaça ou estresse. No entanto, quando ocorre de forma excessiva ou persistente, torna-se patológica, manifestando-se por preocupações excessivas, inquietação, fadiga, dificuldade de concentração e distúrbios do sono. A ansiedade crônica pode interferir significativamente na qualidade de vida e nas atividades diárias.

**Depressão**: A depressão é um transtorno mental caracterizado por humor deprimido, perda de interesse em atividades anteriormente prazerosas, alterações no apetite e no sono, fadiga, sentimento de culpa ou inutilidade e, em casos graves, ideação suicida. É uma condição debilitante que pode afetar todas as esferas da vida do indivíduo, sendo necessária intervenção terapêutica para manejo eficaz.

```python
import plotly.figure_factory as ff

numerical_df = df.select_dtypes(include=['number'])

corr = numerical_df.corr()
fig = ff.create_annotated_heatmap(
    z=corr.values,
    x=list(corr.columns),
    y=list(corr.index),
    annotation_text=corr.round(2).values,
    showscale=True
)
fig.show()
```

Váriaveis de saída que iremos analisar: escala_depressao, burnout_exaustao, escala_ansiedade

```python
target_vars = ['escala_depressao', 'burnout_exaustao', 'escala_ansiedade']

for target in target_vars:
  correlations = corr[target].sort_values(ascending=False)
  correlations = correlations.drop(target)
  top_5 = correlations.head(5)

  print(f"Top 5 variáveis mais correlacionadas com {target}:")
  print(top_5)
  print("\n")
```

# Análise Exploratoria: 5 variáveis mais correlacionadas *Depressão*

```python
vars_interesse = ['escala_ansiedade', 'burnout_exaustao', 'burnout_cinismo', 'terapia_ultimo_ano', 'empatia_afectiva']
df_interesse = df[vars_interesse]
```

**Estatísticas descritivas**

```python
print(df_interesse.describe())
```

**Histogramas**

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, len(vars_interesse), figsize=(20, 4))

for i, var in enumerate(vars_interesse):
  sns.histplot(df_interesse[var], ax=axes[i], kde=True)
  axes[i].set_title(var)
  axes[i].grid(True)

plt.tight_layout()
plt.show()
```

**Boxplots**

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, len(vars_interesse), figsize=(20, 4))

for i, var in enumerate(vars_interesse):
  sns.boxplot(y=df_interesse[var], ax=axes[i])
  axes[i].set_title(var)
  axes[i].grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
terapia_sexo = df.groupby(['sexo_desc', 'terapia_ultimo_ano'])['sexo_desc'].count().unstack().fillna(0)

terapia_sexo.plot(kind='bar', figsize=(10, 6))
plt.title('Terapia no Último Ano por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.legend(title='Terapia no Último Ano', labels=['Não', 'Sim'])
plt.grid(True)
plt.show()
```
